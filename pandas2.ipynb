```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
array=[1,2,4]
plt.plot(array)
```

```python
xpoints=[1,2,3]
ypoints=[5,6,7]
plt.plot(xpoints,ypoints, '*'"r")
plt.show()
```

```python
plt.plot(xpoints,ypoints, marker="*",color="r", ms=18,mec='k')
```

```python
plt.plot(xpoints,ypoints, color="g",mec='k',marker='*',ms=50,mfc='r',linestyle='dashed',linewidth=3)

#marker = shape of points
#color=color of shape  ,ms=size of shape  ,mfc =face color,  linstyle=how line connect points
```

```python
y1=np.array([1,8,6,9])
y2=np.array([5,2,3,7])
plt.plot(y1)
plt.plot(y2)
```

```python
plt.scatter(y1,y2)
```

```python
plt.plot(y1,y2)
```

```python
plt.bar(y1,y2)
```

```python
plt.barh(y1,y2,color="k")
```

```python
x=np.array([3,5,4,6,7,8])
plt.hist(x)
```

```python
plt.pie(y1)
```

```python
y=np.array([200,100,45,87,63])
mylabel=["apple","chery","banana","dates","pineapple"]
plt.pie(y,labels=mylabel)
```

```python
plt.pie(y,labels=mylabel,startangle=0)
```

```python
myexplode=[0.2,0.25,0.1,0.4,0.14]
plt.pie(y,labels=mylabel,startangle=120,explode=myexplode)
```

```python
y=np.array([50,60,80,30,40])
mylabel=["apple","chery","banana","dates","pineapple"]
mycolors=["red","pink","blue","violet","green"]
plt.pie(y,labels=mylabel,colors=mycolors,startangle=50)
```

```python
plt.pie(y)
plt.legend(labels=mylabel)
```
